```python
def BaseForm(m,k):
    A = []
    s = 0
    while m > 0:
      A.append(m%k)
      m = m//k
      B = [i for i,j in enumerate(A)]
    C = [[A[i],B[i]] for i in range(len(A)) if A[i] != 0]
    for i in range(len(C)):
      if C[i][1] >= k:
        m = C[i][1]
        C[i][1] = BaseForm(m,k)
    return C[::-1]
```

```python
def function(m,k):
  A = BaseForm(m,k)
  s = 0
  for i in A:
    if type(i[1]) is int:
      s = s + i[0]*(k+1)**i[1]
    else:
      B = [i]
      while type(i[1]) is not int:
        B.append(i)
        x = i[1]
        i = i[1][0]
      B[len(B)-1] = x
      C = B[::-1]
      a = 0
      for y in C[0]:
        a = a + y[0]*(k+1)**y[1]
      for j in range(1,len(C)):
        C[j][1] = a
        a = C[j][0]*(k+1)**a
      i = C[j]
      s = s + i[0]*(k+1)**i[1]
  return s
```

```python
def Goodstein(m,k):
  if k == 1:
    return m
  else:
    for i in range(2,k+1):
      m = function(m,i) - 1
  return m
```

```python
m = int(input('Enter starting value of goodstein sequence : '))
k = int(input('Enter the number of term which you want : '))
print('The desired term of goodstein sequence is', Goodstein(m,k))
```
