```python
def BaseForm(m,k):
    n = m
    A = []
    s = 0
    while n > 0:
      A.append(n%k)
      n = n//k
      B = [i for i,j in enumerate(A)]
    C = [(A[i],B[i]) for i in range(len(A)) if A[i] != 0]
    for i in range(len(C)):
      if C[i][1] >= k:
        C[i] = list(C[i])
        C[i][1] = BaseForm(C[i][1],k)
        C[i] = tuple(C[i])
    return C
```

```python
def function(m,k):
    if type(m) is int:
        return m
    elif type(m) is tuple:
        return m[0]*k**function(m[1], k)
    elif type(m) is list:
        return sum([function(i, k) for i in m])
```

```python
def Goodstein(m,k):
  if k == 1:
    return m
  else:
    for i in range(2,k+1):
      A = BaseForm(m,i)
      m = function(A, i+1) - 1
    return m
```

```python
m = int(input('Enter starting value of goodstein sequence : '))
k = int(input('Enter the number of term which you want : '))
print('The desired term of goodstein sequence is', Goodstein(m,k))
```
